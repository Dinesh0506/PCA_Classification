```python
#importing the libraries
import pandas as pd
from sklearn.model_selection import train_test_split
import time
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
from sklearn.feature_selection import RFE
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
import pickle
import matplotlib.pyplot as plt

# Applying PCA
def apply_pca(X_train, X_test, n_components=2):
    pca = PCA(n_components=n_components)
    X_train_pca = pca.fit_transform(X_train)
    X_test_pca = pca.transform(X_test)
    explained_variance = pca.explained_variance_ratio_
    return X_train_pca, X_test_pca, explained_variance

# Split and standardize data
def Split_scalar(indep_X, dep_Y):
    X_train, X_test, y_train, y_test = train_test_split(indep_X, dep_Y, test_size=0.25, random_state=0)
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)
    return X_train, X_test, y_train, y_test

# Model accuracy validation
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
def cm_prediction(classifier, X_test, y_test):
    y_pred = classifier.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)
    Accuracy = accuracy_score(y_test, y_pred)
    report = classification_report(y_test, y_pred)
    return classifier, Accuracy, report, cm



# Logistic regression model
def logistic(X_train, y_train, X_test, y_test):
    classifier = LogisticRegression(random_state=0)
    classifier.fit(X_train, y_train)
    return cm_prediction(classifier, X_test, y_test)

# SVM Linear model
def svm_linear(X_train, y_train, X_test, y_test):
    from sklearn.svm import SVC
    classifier = SVC(kernel='linear', random_state=0)
    classifier.fit(X_train, y_train)
    return cm_prediction(classifier, X_test, y_test)

# SVM Non-linear model
def svm_NL(X_train, y_train, X_test, y_test):
    from sklearn.svm import SVC
    classifier = SVC(kernel='rbf', random_state=0)
    classifier.fit(X_train, y_train)
    return cm_prediction(classifier, X_test, y_test)

# Naive Bayes model
def Navie(X_train, y_train, X_test, y_test):
    from sklearn.naive_bayes import GaussianNB
    classifier = GaussianNB()
    classifier.fit(X_train, y_train)
    return cm_prediction(classifier, X_test, y_test)

# KNN model
def knn(X_train, y_train, X_test, y_test):
    from sklearn.neighbors import KNeighborsClassifier
    classifier = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
    classifier.fit(X_train, y_train)
    return cm_prediction(classifier, X_test, y_test)

# Decision Tree model
def Decision(X_train, y_train, X_test, y_test):
    from sklearn.tree import DecisionTreeClassifier
    classifier = DecisionTreeClassifier(criterion='entropy', random_state=0)
    classifier.fit(X_train, y_train)
    return cm_prediction(classifier, X_test, y_test)

# Random Forest model
def random(X_train, y_train, X_test, y_test):
    from sklearn.ensemble import RandomForestClassifier
    classifier = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=0)
    classifier.fit(X_train, y_train)
    return cm_prediction(classifier, X_test, y_test)

# Compile results in a DataFrame
def selectk_Classification(acclog, accsvml, accsvmnl, accknn, accnav, accdes, accrf):
    dataframe = pd.DataFrame(index=[1,2,3,4,5], columns=['Logistic', 'SVMl', 'SVMnl', 'KNN', 'Navie', 'Decision', 'Random'])
    dataframe['Logistic'] = acclog
    dataframe['SVMl'] = accsvml
    dataframe['SVMnl'] = accsvmnl
    dataframe['KNN'] = accknn
    dataframe['Navie'] = accnav
    dataframe['Decision'] = accdes
    dataframe['Random'] = accrf
    return dataframe

# Load the dataset
dataset1 = pd.read_csv("prep.csv")
df2 = pd.get_dummies(dataset1, drop_first=True)

# Split data into features and target variable
indep_X = df2.drop('classification_yes', axis=1)
dep_Y = df2['classification_yes']

# Apply scaling and PCA
X_train, X_test, y_train, y_test = Split_scalar(indep_X, dep_Y)

# Initialize lists for model accuracies
acclog, accsvml, accsvmnl, accknn, accnav, accdes, accrf = [], [], [], [], [], [], []
for n_components in range(1, 6):
    print(f"\nApplying PCA with {n_components} components:")
    X_train_pca, X_test_pca, explained_variance = apply_pca(X_train, X_test, n_components=2)
    # Calculate accuracy for each model
    classifier, Accuracy, report, cm = logistic(X_train_pca, y_train, X_test_pca, y_test)
    acclog.append(Accuracy)
    
    classifier, Accuracy, report, cm = svm_linear(X_train_pca, y_train, X_test_pca, y_test)
    accsvml.append(Accuracy)
    
    classifier, Accuracy, report, cm = svm_NL(X_train_pca, y_train, X_test_pca, y_test)
    accsvmnl.append(Accuracy)
    
    classifier, Accuracy, report, cm = knn(X_train_pca, y_train, X_test_pca, y_test)
    accknn.append(Accuracy)
    
    classifier, Accuracy, report, cm = Navie(X_train_pca, y_train, X_test_pca, y_test)
    accnav.append(Accuracy)
    
    classifier, Accuracy, report, cm = Decision(X_train_pca, y_train, X_test_pca, y_test)
    accdes.append(Accuracy)
    
    classifier, Accuracy, report, cm = random(X_train_pca, y_train, X_test_pca, y_test)
    accrf.append(Accuracy)

# Store results in a DataFrame
result = selectk_Classification(acclog, accsvml, accsvmnl, accknn, accnav, accdes, accrf)
print(result)
```
